```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
x_train[0]
```

```python
x_train=x_train.reshape(x_train.shape[0],28,28 , 1)
```

```python
x_test=x_test.reshape(x_test.shape[0],28,28 , 1)
```

```python
from tensorflow.keras import layers , models
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Conv2D, MaxPool2D,Flatten,Dense,Dropout
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer="adam" , loss="sparse_categorical_crossentropy" , metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
test_loss , test_acc=model.evaluate(x_test , y_test)
print(test_acc ," is the test accuracy for test data")
```

```python
import random
import numpy as np


class_names=['T-shirt' , 'Trouser' , 'Pullover' , 'Dress' , 'Coat' ,
             'Sandal' ,'Shirt' ,'Sneaker' ,'Bag' ,'Ankle Boot']
index =random.randint(0 , len(x_test)-1)
random_image=x_test[index]


true_label=y_test[index]

image_for_prediction=np.expand_dims(random_image , axis=0)
predictions= model.predict(image_for_prediction)
predicted_label_index=np.argmax(predictions[0])

predicted_label_name=class_names[predicted_label_index]

true_label_name=class_names[true_label]


plt.imshow(random_image.reshape(28,28) , cmap=plt.cm.binary)
plt.title(f"Predicted Label: {predicted_label_name} \n True Label: {true_label_name}")
plt.show()
```
